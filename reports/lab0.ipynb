# CSV File(s)

```python
import csv

# Define the file name
file_name = 'sample_data.csv'

# Define the header and some sample data
header = ['Name', 'Age', 'City']
data = [
    ['Alice', 30, 'New York'],
    ['Bob', 24, 'Los Angeles'],
    ['Charlie', 35, 'Chicago']
]

# Write the data to the CSV file
with open(file_name, 'w', newline='') as csvfile:
    csv_writer = csv.writer(csvfile)
    csv_writer.writerow(header) # Write the header
    csv_writer.writerows(data)  # Write the data rows

print(f"CSV file '{file_name}' created successfully.")

# You can verify its content by running:
# !cat sample_data.csv
```

# JSON File(s)

```python
import json

# Define the file name
json_file_name = 'sample_data.json'

# Define some sample data
data = [
    {'Name': 'Alice', 'Age': 30, 'City': 'New York'},
    {'Name': 'Bob', 'Age': 24, 'City': 'Los Angeles'},
    {'Name': 'Charlie', 'Age': 35, 'City': 'Chicago'}
]

# Write the data to the JSON file
with open(json_file_name, 'w') as jsonfile:
    json.dump(data, jsonfile, indent=4)

print(f"JSON file '{json_file_name}' created successfully.")
```

# Reading File(s) w/ Numpy

### TXT File(s)

```python
# Define the file name
txt_file_name = 'sample_data.txt'

# Define some sample content
content = "This is a sample text file.\nIt contains multiple lines of text.\nHello, world!"

# Write the content to the text file (if not already created)
with open(txt_file_name, 'w') as txtfile:
    txtfile.write(content)

print(f"Text file '{txt_file_name}' ensured to be created.")

# Now, read the content of the text file
with open(txt_file_name, 'r') as txtfile:
    read_content = txtfile.read()

print(f"\nContent of '{txt_file_name}':\n{read_content}")
```

### JSON File(s)

```python
import json
import numpy as np

json_file_name = 'sample_data.json'

# Step 1: Read the JSON file into a Python list of dictionaries
with open(json_file_name, 'r') as jsonfile:
    data_from_json = json.load(jsonfile)

print(f"Data read from {json_file_name} (Python list of dicts):")
print(data_from_json)

# Step 2: Convert the list of dictionaries to a structured NumPy array
# First, define the dtype for the structured array based on the JSON structure
dtype = [('Name', 'U50'), ('Age', 'i4'), ('City', 'U50')]

# Create an empty structured array
numpy_json_array = np.empty(len(data_from_json), dtype=dtype)

# Populate the structured array
for i, item in enumerate(data_from_json):
    numpy_json_array[i]['Name'] = item['Name']
    numpy_json_array[i]['Age'] = item['Age']
    numpy_json_array[i]['City'] = item['City']

print(f"\nContent of {json_file_name} as a structured NumPy array:")
print(numpy_json_array)
print("Data type of the NumPy array:", numpy_json_array.dtype)
```

### CSV File(s)

```python
import numpy as np

csv_file_name = 'sample_data.csv'

# Use np.genfromtxt to read the CSV file
# delimiter=',' specifies that values are separated by commas
# skip_header=1 skips the first row, which is assumed to be the header
# dtype=None allows NumPy to infer the data types of columns
# encoding='utf-8' is good practice for text files
csv_data = np.genfromtxt(csv_file_name, delimiter=',', skip_header=1, dtype=None, encoding='utf-8')

print(f"Content of {csv_file_name} using NumPy:")
print(csv_data)

# If you need to access columns by name, you can convert to a structured array
# You would typically define dtype with names explicitly for more complex CSVs
# For simple cases like this, it reads as a 2D array or structured array if dtypes vary

# To access header separately if needed:
# with open(csv_file_name, 'r') as f:
#     header = f.readline().strip().split(',')
# print("Header:", header)
```

### Initializing SparkSession

```python
from pyspark.sql import SparkSession

# Create a SparkSession
spark = SparkSession.builder.appName("ReadFiles").getOrCreate()

print("SparkSession created successfully.")
```

### Reading `sample_data.csv` with PySpark

```python
csv_file_name = 'sample_data.csv'
# Read CSV file, inferSchema to detect data types, and header to use the first row as column names
spark_csv_df = spark.read.csv(csv_file_name, header=True, inferSchema=True)

print(f"Content of {csv_file_name} using PySpark:")
spark_csv_df.show()
spark_csv_df.printSchema()
```

### Reading `sample_data.json` with PySpark

```python
json_file_name = 'sample_data.json'
# Read JSON file, inferSchema to detect data types, and specify multiLine=True for JSON arrays
spark_json_df = spark.read.json(json_file_name, multiLine=True, inferSchema=True)

print(f"Content of {json_file_name} using PySpark:")
spark_json_df.show()
spark_json_df.printSchema()
```
